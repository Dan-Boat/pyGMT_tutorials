### Fames, ticks and labels

This session shows how to set the various frame funtionalities, set ticks and add labels

```python
import pygmt as gmt
```

```python
# set with defualt frame "f"
fig1 = gmt.Figure()
fig1.coast(shorelines="1/0.5p", region=[-180, 180, -60, 60], projection="M25c")
fig1.basemap(frame="f")
fig1.show()
```

```python
# adding frame="a" set the ticks for the automatic frames and adding "g" display gridlines
# The title can also pass to the basemap with the "+t---" parameter to frame 
fig2 = gmt.Figure()
fig2.coast(shorelines="1/0.5p", region=[-180, 180, -60, 60], projection="M25c")
fig2.basemap(frame="ag")
fig2.show()
```

```python
# country codes can also be used to set the region
fig3 = gmt.Figure()
fig3.coast(shorelines="1/0.5p", region="TT", projection="M25c")
fig3.basemap(frame=["a", '+t"Trinidad and Tobago"'])
fig3.show()
```

```python
# Normal axis can also be defined
fig = gmt.Figure()
fig.basemap(
    region=[0, 10, 0, 20],
    projection="X10c/8c",
    # Plot axis, tick marks, and axis labels on the west/left and south/bottom
    # axes
    # Plot axis and tick marks on the north/top and east/right axes
    frame=["WSne", "xaf+lx-axis", "yaf+ly-axis"],
)
fig.show()
```
