### This example basically checks the installation of pygmt with a typical hello work example..funny right!

#### To install the package and its dependencies on a new environment, the following steps can be followed;

###### 1. Create the a new environment with the name pygmt and install some dependencies
* conda create --name pygmt python=3.9 numpy pandas xarray netcdf4 packaging gmt

##### 2. activate the conda env to install the module using conda forge
* conda activate pygmt
* conda install pygmt

##### 3. check if everything installed by running the following sections

```python
#Trying if the installation process worked successfully
import pygmt
pygmt.show_versions()
```

```python
# creating first land-sea mask with normal frame and EckertI IV projection
fig = pygmt.Figure()
fig.coast(region="g", shorelines=1, land="lightgreen", water="lightblue", projection="G10/52/12c", frame=["afg", "+tMy Hello, World pyGMT"])
fig.show()
```

```python
fig = pygmt.Figure()
# Use region "d" to specify global region (-180/180/-90/90)
fig.coast(region="d", projection="Kf12c", land="ivory", water="bisque4", frame=["afg", "+tMy Hello, World pyGMT"])
fig.show()
```
